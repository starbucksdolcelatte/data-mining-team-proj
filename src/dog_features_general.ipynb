```python
import csv
from urllib.request import urlopen, Request
from bs4 import BeautifulSoup as bs

# newline = ''으로 함으로써 empty line 이 사라진다.
with open('fundamental2.csv', 'w', newline='') as outfile:
    writer = csv.writer(outfile)
    
    url = r'http://finviz.com/quote.ashx?t={}'.format("NVDA".lower())
    req = Request(url)
    html = urlopen(req).read()
    soup = bs(html, 'lxml')
    names = soup.findAll(class_='snapshot-td2-cp')
    datas = soup.findAll(class_='snapshot-td2')
    print(writer)
    try :
        for i in range(len(names)):
            name = names[i].getText()
            print(name)
            data = datas[i].getText()
            print(data)
            writer.writerow(("NVDA", name,data))
    finally :
        outfile.close()
```
